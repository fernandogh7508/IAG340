# Limpieza e ingeniería de características

Relleno de nulos encoding y creación de variables nuevas

```python
#crear copia para trabajar
data= df.copy()
```

```python
# Extraer el título
data['title']=data['Name'].str.extract(r'(Mr\.|Mrs\.|Miss\.)')
#contar cuántos hay de cada uno
data['title'].value_counts()
```

```python
# Age:rellenar con mediana
if data['Age'].isnull().sum()>0:
  data['Age'].fillna(data['Age'].median(),inplace=True)
```

```python
data.info()
```

```python
data.head()
```

```python
#Embarked: rellenar con la moda si hay nulo
if 'Embarked' in data.columns:
  data['Embarked'].fillna(data['Embarked'].mode()[0],inplace=True)
data.info()
```

```python
data.head()
```

```python
#Encoding: sex y Embarked
if 'Sex' in data.columns:
  data['Sex']=data['Sex'].map({'male':0,'female':1})


if 'Embarked' in data.columns:
  data['Embarked']=data['Embarked'].map({'S':0,'C':1,'Q':2})



```

```python
data.head()
```

# Preparación para modelado

```python
features=[]
for col in ['Pclass','Sex','Age','Embarked']:
  if col in data.columns:
    features.append(col)
x=data[features]
y=data['Survived']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test,=train_test_split(x,y,test_size=0.2,random_state=42)
print("features usadas",features)
print('Train shape:',x_train.shape,'test shape:',x_test.shape)
```

```python
#ver primeras filas
print( '\n Primera filas de x_train')
print(x_train.head())
```

```python
print('\n Primera filas de x_test')
print(x_test.head())
```

```python
#verificar que no haya duplicados entre train y test
duplicados=x_train.index.intersection(x_test.index)
print('filas duplicadas',len(duplicados))
if len(duplicados)>=0:
  print('Hay duplicados entre train y test')
```

```python
#Entrenar el modelo
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
model = DecisionTreeClassifier(random_state=42)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(" Accuracy =",accuracy)
```
